```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/sales_by_price.csv')

df
```

```python
sns.set_theme(style= 'whitegrid', palette= 'bright')
sns.histplot(data= df, x='num_sold', bins= 3)
```

```python
dftest = pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/discount_items.csv')
dftest
```

```python
sns.set_theme(style= 'whitegrid', palette = 'bright')
sns.histplot(data = dftest, x='price', bins= 3)
plt.savefig('Test_visual.png',dpi='figure',  format= 'png')
plt.show()
```

```python
sns.set_theme(style= 'darkgrid', palette = 'dark')
sns.histplot(data = dftest, x='price', bins= 50)
```
