```python
import pandas as pd
import matplotlib.pyplot as plt

pathway ='https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/salary_data.csv'
df = pd.read_csv(pathway)
df
```

```python
plt.pie([df['Assoc'].sum(), df['Spv'].sum(), df['Mgr'].sum()], labels=['Associates', 'Supervisors', 'Managers'], autopct='%1.1f%%', colors=['yellowgreen','gold','lightskyblue'])
#plt.legend(labels=['Associates', 'Supervisors', 'Managers'])
plt.show()
```
