```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-science-lectures/master/emp_avg_data.csv')

df
```

```python
sns.barplot(x='avg_emp_rating', y='avg_emp_tenure_in_years', data=df, hue='emp_break_room')
```

```python
sns.set_theme(style='darkgrid', palette='dark')
sns.barplot(x='avg_emp_rating', y='avg_emp_tenure_in_years', data=df, hue='emp_break_room')
plt.show()
```
